# 2. Explain the difference between:
=> Mean,Median,Mode
=> Range,Variance,Standard Deviation

# Answer.

1. Mean, Median, Mode

Mean
The average of all values.
Calculated by adding all values and dividing by the number of values.
Example: Average household income.

Median
The middle value when data is arranged in order.
Less affected by extreme values (outliers).
Example: Middle age of household heads.

Mode
The most frequently occurring value.
Can be more than one mode or none.
Example: Most common education level.

2. Range, Variance, Standard Deviation

Range
Difference between the highest and lowest values.
Shows how spread out the data is.
Example: Max income âˆ’ Min income.

Variance
Measures how far values are spread from the mean (average).
Calculated using squared differences from the mean.
Expressed in squared units.

Standard Deviation
The square root of variance.
Shows spread of data in the same units as the data.
Easier to interpret than variance.
