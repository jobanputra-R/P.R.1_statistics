```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from  scipy.stats import kurtosis
from  scipy.stats import skew

df=pd.read_csv("household_data.csv")
#3. Types of data
print(df.dtypes)
```

```python
#4.Central tendency
income=df["Household_Income"]
age=df["Age_of_Household_Head"]
sort_income=income.sort_values()
sort_age=age.sort_values()
mean_age=age.mean()
mean_income=income.mean()
median_age=sort_age.median()
median_income=sort_income.median()
mode_age=age.mode()[0]
mode_income=income.mode()[0]

print("Mean of Income is :",mean_income)
print("Mean of Age is :",mean_age)
#the average income is 68313
# the average age is 48 
print("Median of Income is :",median_income)
print("Median of Age is :",median_age)
# Middle income is 64972.5
# middle age is 50
print("Mode of Income is :",mode_income)
print("Mode of Age is :",mode_age)
#most repeated income is 15224
# most repeated age is 63
```

```python
#5.Measure of dispersion 
range_income=income.max()-income.min()
print("The range of income is :",range_income)
variance_income=income.var()
print("variance of income is :",variance_income)
standard_deviation=income.std()
print("Standard deviation of income is :",standard_deviation)

# finding iqr
q1=np.percentile(income,25)
q2=median_income
q3=np.percentile(income,75)
iqr=q3-q1
print(f"Q1 is {q1} Q2 is {q2} Q3 is {q3}")
print("IQR is :",iqr)
```

```python
#6.Distribution 
plt.figure(figsize=(10,7))
sns.histplot(income,kde=True,bins=10)
plt.axvline(mean_income,color='red',linestyle="--",label="Mean")
plt.axvline(median_income,color='yellow',linestyle="--",label="Median")
plt.title("Household Income")
plt.show()
```

```python
#calculating skewness and kurtosis of the income data 
skewness_income=skew(income)
kurtosis_income=kurtosis(income,fisher=False)
print("Skewness of income is :",skewness_income)
print("Kurtosis of Household income is :",kurtosis_income)
```

```python
# 7 Data Categorize (visualization histogram)
plt.figure()
sns.histplot(data=df,x=income,hue="Education_level",bins=20)
plt.title("Household Income Across Education levels")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df,x="Education_level",y="Family_Size")
plt.title("Family size by education level")
plt.tight_layout()
plt.show()
```
